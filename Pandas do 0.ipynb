# .loc, um metodo muito importante

    - pegar 1 linha
    - pegar linhas de acordo com alguma condição
    - pegar linhas e colunas especificadas
    - pegar 1 valor especifico

```python
# pegar 1 linha
display(vendas_df.loc[1:5])
```

```python
# pegar linha que correspondem a uma condição
display(vendas_df.loc[vendas_df['ID Loja'] == "Norte Shopping"])
```

```python
# pegar varias linhas e colunas usando loc
# vendas_df[linhas,colunas]
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == "Norte Shopping", ["ID Loja", "Produto", "Quantidade"]]
display(vendas_norteshopping_df)
```

```python
#pegar um valor especifico

print(vendas_df.loc[23214,"Produto"])
```

# Adicionar 1 coluna

```python
# A partir de uma coluna que existe
vendas_df['Comissão'] = vendas_df["Valor Final"] *0.05
display(vendas_df)

# Criar uma coluna com valor padrão
vendas_df.loc[:,"Imposto"] = 0
display(vendas_df)
```

# Adicionar linhas 
    - Linhas de um complemento da base de dados
    

```python
vendas_dez_df = pd.read_excel("Vendas_Dez.xlsx")
vendas_df = vendas_df.append(vendas_dez_df)
display(vendas_df)
```

# Excluir  linhas e colunas
    - axis = 0 --> excluir linha
    - axis = 1 --> excluir coluna
    

```python
vendas_df = vendas_df.drop("Imposto", axis=1)
vendas_df
```

# Valores Vazios

    - Deletar linhas/colunas vazias
    - Deletar linhas que possuem valores vazios
    - Preencher valores vazios (média e ultimo valor)

```python
# deletar linhas e colunas completamente vazios
vendas_df =vendas_df.dropna(how="all")

# deletar linhas que possuem pelo menos 1 valor vazio
vendas_df = vendas_df.dropna()

# preencher valores vazios
    #-prencher com a media da coluna
vendas_df["Comissão"] = vendas_df["Comissão"].fillna(vendas_df["Comissão"].mean())
display(vendas_df)

    # preencher com o ultimo valor
vendas_df = vendas_df.ffill()
```

# Calculcar indicadores
    - Groupby
    - Value Count

```python
#Value_counts
transações_lojas = vendas_df['ID Loja'].value_counts()
transações_lojas

#Groupby

faturamento_produto = vendas_df[["Produto", "Valor Final"]].groupby("Produto").sum()
faturamento_produto
```

# Mesclar 2 dataframes( Procurar informações de um dataframe em outros)


```python
gerentes_df = pd.read_excel("Gerentes.xlsx")

vendas_df = vendas_df.merge(gerentes_df)
vendas_df
```
